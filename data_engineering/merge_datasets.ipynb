```python
import pandas as pd
```

```python
#load the two datasets to work with

translated = pd.read_csv('../data/mediacloud-translated-completed_taba.csv')
orientation = pd.read_csv('../data/media_cloud-political_orientation.csv')
```

```python
len(set(translated["id"]) & set(orientation["id"]))
```

```python
#count duplicates ids in both datasets
print(f"Duplicates in translated: {translated['id'].duplicated().sum()}")
print(f"Duplicates in orientation: {orientation['id'].duplicated().sum()}")
```

```python
#drop duplicates based on id in both datasets
translated = translated.drop_duplicates()
orientation = orientation.drop_duplicates()
```

```python
# Set index to ID for alignment
translated_idx = translated.set_index("id")
orientation_idx = orientation.set_index("id")

# Replace titles where translated_title exists
orientation_idx.loc[
    translated_idx.index, "title"
] = translated_idx["translated_title"]

```

```python
# Merge translated_title into orientation
orientation = orientation.merge(
    translated[["id", "translated_title"]],
    on="id",
    how="left"
)

# Replace title with translated_title where available
orientation["title"] = orientation["translated_title"].combine_first(orientation["title"])

# Drop the temporary column
orientation = orientation.drop(columns="translated_title")
```

```python
#save the merged dataset
orientation.to_csv('../data/orientaion_english.csv', index=False)
```
